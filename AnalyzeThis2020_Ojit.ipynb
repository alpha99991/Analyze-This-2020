```
d_col = ['count_accts', 'cm_age', 'flag_top_ed_spender','flag_cust_fee_paid_6m','pre6m_cust_roc_cnt','pre6m_cust_non_disc_cnt','pre6m_cust_disc_cnt', 'pre6m_cust_online_cnt','pre6m_cust_travel_cnt','pre6m_cust_retail_cnt', 'pre6m_cust_myca_active', 'pre6m_cust_mob_logins']
```

```
for i in d_col:
    dfc[i] = dfc[i].apply(f)
```

```
dfc.head()
```

```
col = ['Cust_tenure','pre6m_spend_active_ind','highly_utilized_ind','min_pay_ind','paid_in_full_ind','direct_debit_ind','referrals', 'spillover', 'self_accts',	'Customer Low Quality indicator', 'cnsumr_lend_actv_cust_cnt', 'sow_tot_revol_cnt',	'sow_tot_trans_cnt']
```

```
for i in col:
    dfc[i] = dfc[i].apply(f)
```

```
dfc.head()
```

```
dfc['cnsumr_chrg_actv_cust_cnt'] = dfc['cnsumr_chrg_actv_cust_cnt'].apply(f)
```

```
dfc.dtypes
```

```
ocol = ['pre6m_total_mc_trs', 'acq_sub_chn', 'acq_type_grp', 'fee_type_grp']
count  = 0  
l  = []  
for i in dfc[ocol[0]]:
    i = list(i)
    if(i.count('(')!=0):
        l.append(int(("".join(i[2:-1])).replace(',', '')))
    else:
        l.append(int(("".join(i[1:-1])).replace(',', '')))     
            
                        
                       
                      
```

```
len(l)
```

```
dfc[ocol[0]] = l
```

```
dfc.head()
```

```
dfc.dtypes
```

```
dfc['acq_type_grp'].value_counts()
```

```
dfc['acq_type_grp'] = dfc['acq_type_grp'].map({'external':0, 'spillover':1, 'others':2})
```

```
dfc['acq_type_grp']
```

```
dfc['fee_type_grp'].value_counts()	
```

```
dfc['fee_type_grp'] = dfc['fee_type_grp'].map({'no_fee':0, 'full_fee':1, 'reduced_fee':2, 'unkonown':3})
```

```
dfc[ocol[1]] = dfc[ocol[1]].astype('category')
```

```
dfc[ocol[1]] = dfc[ocol[1]].cat.codes
```

```
dfc.dtypes
```

```
dfc['fee_type_grp'] = dfc['fee_type_grp'].astype('float32')
dfc['cdss_most_rcnt_prob'] = dfc['cdss_most_rcnt_prob'].astype('float32')
dfc['cbr_3_score'] = dfc['cbr_3_score'].astype('float32')
dfc['cnsumr_lend_tot_util_ratio'] = dfc['cnsumr_lend_tot_util_ratio'].astype('float32')
dfc['sow_revol_avg_paydown_pct'] = dfc['sow_revol_avg_paydown_pct'].astype('float32')
```

```
dfc['fee_type_grp'].fillna(dfc['fee_type_grp'].median(), inplace = True)
```

```
y_pred = clf.predict(x_test)
```

```
from sklearn.metrics import f1_score, accuracy_score
```
